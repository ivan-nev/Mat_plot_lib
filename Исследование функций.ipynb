Исследование функций
$$\LARGE f(x)=sin(\frac{x}{5})\cdot e^{\frac{x}{10}}+5\cdot e^{-\frac{x}{2}}$$
$$\LARGE g(x)=1-x^{3}+x^{5}$$
$$\LARGE h(x)={(5\cdot sin^{2}(e^{0.8x})\cdot x^{2}})/({e^{\frac{x^{2}}{2}}})$$



```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


def g(x):
    return 1 - x ** 3 + x ** 5


def h(x):
    return (5 * (np.sin(np.exp(0.8 * x)) ** 2 * x ** 2)) / (np.exp(x ** 2 / 2))


def extremum(func, x: np.ndarray, axes, color='ro'):
    derivative = np.gradient(func(x), x)
    extrema_indices = np.where(np.diff(np.sign(derivative)))
    extrema_x = x[extrema_indices]
    extrema_y = func(x[extrema_indices])
    axes.plot(extrema_x, extrema_y, color)
    axes.tick_params(axis='x', rotation=70)
    axes.set_xticks(np.around(np.concatenate([np.linspace(x[0], x[-1], 2), extrema_x]), 2))
    axes.set_yticks(np.around(np.concatenate([extrema_y, ]), 2))


# plt.plot(extrema_x, extrema_y, 'bo')


fig, ax = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle(r'$f(x)=sin(\frac{x}{5})\cdot e^{\frac{x}{10}}+5\cdot e^{-\frac{x}{2}}$ точки min, max на диапазоне',
             fontsize=20)
xs = [np.linspace(-50, 600, 1000), np.linspace(0, 30, 1000), np.linspace(30, 95, 1000)]
for i, x in enumerate(xs):
    ax[i].plot(x, f(x), 'b', label=f'от {x[0]} до {x[-1]}')
    ax[i].legend()
    ax[i].grid(which='major', linestyle=':', color='#aaa')
    ax[i].set_xlabel('Ось Х')
    ax[i].set_ylabel('Значение функций')
ax[0].set_ylim(-10 ** 10, 10 ** 10)
extremum(f, xs[1], ax[1], )
extremum(f, xs[2], ax[2])
plt.show()
```

Выводы
- при x -> -inf f(x)-> +inf
- при х=0 f(x)=5
- при x -> inf f(x)-> nan
- период 2Pi/5

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle(r'$g(x)=1-x^{3}+x^{5}$ точки min, max на диапазоне', fontsize=20)
xs = [np.linspace(-1000, 1000, 100), np.linspace(-5, 5, 100), np.linspace(-1.2, 1.1, 5000)]
for i, x in enumerate(xs):
    ax[i].plot(x, g(x), 'r', label=f'от {x[0]} до {x[-1]}')
    ax[i].legend()
    ax[i].grid(which='major', linestyle=':', color='#aaa')
    ax[i].set_xlabel('Ось Х')
    ax[i].set_ylabel('Значение функций')
extremum(g, xs[2], ax[2], 'go')
ax[2].set_xticks(np.concatenate([ax[2].get_xticks(), [0]]))
ax[2].set_yticks(np.concatenate([ax[2].get_yticks(), [g(0)]]))
ax[2].plot(0, g(0), 'bo')
plt.show()
```

Выводы
- при x -> -inf g(x)-> -inf
- при х=0 g(x)=1
- при x -> inf g(x)-> inf

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle(r'$h(x)={(5\cdot sin^{2}(e^{0.8x})\cdot x^{2}})/({e^{\frac{x^{2}}{2}}})$ точки min, max на диапазоне',
             fontsize=20)
xs = [np.linspace(-1, 3.15, 1000), np.linspace(3, 4, 1000), np.linspace(-1.3, 1.3, 1000)]
for i, x in enumerate(xs):
    ax[i].plot(x, h(x), 'g', label=f'от {x[0]} до {x[-1]}')
    ax[i].legend()
    ax[i].grid(which='major', linestyle=':', color='#aaa')
    ax[i].set_xlabel('Ось Х')
    ax[i].set_ylabel('Значение функций')
    extremum(h, x, ax[i], 'bo')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings

warnings.filterwarnings('ignore')


def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)


def g(x):
    return 1 - x ** 3 + x ** 5


def h(x):
    return (5 * (np.sin(np.exp(0.8 * x)) ** 2 * x ** 2)) / (np.exp(x ** 2 / 2))


functions = [f, g, h]
legends = [r'$f(x)=sin(\frac{x}{5})\cdot e^{\frac{x}{10}}+5\cdot e^{-\frac{x}{2}}$', r'$g(x)=1-x^{3}+x^{5}$',
           r'$h(x)={(5\cdot sin^{2}(e^{0.8x})\cdot x^{2}})/({e^{\frac{x^{2}}{2}}})$']
colors = ['blue', 'red', 'green']
# arias = [np.linspace(-2, 0, 500), np.linspace(0.25, 1.5, 500), np.linspace(200, 350, 10000), np.linspace(10, 70, 10000)]
# arias = [np.linspace(-2, 0, 1000), np.linspace(0, 4, 2000), np.linspace(200, 300, 10000), np.linspace(4, 70, 10000)]
arias = [np.linspace(-2, 0, 1000), np.linspace(0, 4, 2000), np.linspace(4, 70, 10000), np.linspace(200, 300, 10000)]

fig, axes = plt.subplot_mosaic([[0, 1], [2, 3]], figsize=(20, 20), dpi=100)
fig.suptitle('Нахождение точек пересечения', fontsize=20)

# axes[0].set_ylim(-10, 10)
# axes[3].set_ylim(-50, 50)
# axes[3].set_yscale("symlog")
for i, aria in enumerate(arias):
    for function, legend, color in zip(functions, legends, colors):
        axes[i].plot(aria, function(aria), label=legend, color=color)

    axes[i].set_title(f'Диапазон от {aria[0]} до {aria[-1]}')
    idx_gh = np.argwhere(np.diff(np.sign(g(aria) - h(aria)))).flatten()
    idx_fg = np.argwhere(np.diff(np.sign(f(aria) - g(aria)))).flatten()
    idx_fh = np.argwhere(np.diff(np.sign(f(aria) - h(aria)))).flatten()

    axes[i].plot(aria[idx_gh], g(aria[idx_gh]), 'bo')
    axes[i].plot(aria[idx_fg], g(aria[idx_fg]), 'go')
    axes[i].plot(aria[idx_fh], f(aria[idx_fh]), 'ro')

    dot = np.concatenate([idx_gh, idx_fg, idx_fh])
    y_dot = np.concatenate([h(aria[dot]), g(aria[dot])])
    axes[i].set_xticks(np.around(aria[dot], 2))
    axes[i].set_yticks(np.around(y_dot, 2))
    axes[i].tick_params(axis='x', rotation=70)

    axes[i].set_xlabel('Ось Х')
    axes[i].set_ylabel('Значение функций')
    axes[i].legend(fontsize=15)
    axes[i].grid(which='major', linestyle='--', color='#aaa')

axes[1].set_ylim(-0.5, 5)
axes[2].set_ylim(-500, 500)
axes[3].set_ylim(-1 * 10 ** 12, 2.5 * 10 ** 12)
# plt.savefig('123.pdf')
plt.show()
```
