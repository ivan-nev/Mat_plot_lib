## Задача №1

### Часть 1

Построить в общих осях графики для:

- $\alpha = 1, \beta = 1$

- $\alpha = 2, \beta = 1$

- $\alpha = 1, \beta = 2$

```python
import  matplotlib.pyplot  as  plt
import numpy as np

xp = np.arange(0.1, 4, 0.02)
xn = -1 * xp
a = 1, 2, 1
b = 1, 1, 2

yp = [(xp**bi + ai**bi) / xp**bi for ai, bi in zip(a, b)]
yn = [(xn**bi + ai**bi) / xn**bi for ai, bi in zip(a, b)]

plt.figure(figsize=(13,8), dpi=120)
plt.plot(xp, yp[0], 'r', label='$f(x;1,1) = \\frac{x+1}{x}$')
plt.plot(xn, yn[0], 'r')
plt.plot(xp, yp[1], 'g', label='$f(x;2,1) = \\frac{x+2}{x}$')
plt.plot(xn, yn[1], 'g')
plt.plot(xp, yp[2], 'b', label='$f(x;1,2) = \\frac{x^2+1}{x^2}$')
plt.plot(xn, yn[2], 'b')
plt.title('Task 1, part 1', fontsize=18)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f(x;\\alpha, \\beta)$', fontsize=15)
plt.legend(loc = 5, fontsize=15)
plt.rc('grid', linestyle=':', color='gray', linewidth= 1)
plt.grid(True)
plt.savefig('Task-1_part-1.svg')
plt.show()
```

```python
xn = -1 * np.arange(0.1, 5, 0.02)
a = 1, 2, 1
b = 1, 1, 2

yn = [(xn**bi + ai**bi) / xn**bi for ai, bi in zip(a, b)]

plt.figure(figsize=(13,8), dpi=120)
plt.plot(xn, yn[0], 'r', label='$f(x;1,1) = \\frac{x+1}{x}$')
plt.plot(xn, yn[1], 'g', label='$f(x;2,1) = \\frac{x+2}{x}$')
plt.plot(xn, yn[2], 'b', label='$f(x;1,2) = \\frac{x^2+1}{x^2}$')
plt.title('Task 1, part 3', fontsize=18)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f(x;\\alpha, \\beta)$', fontsize=15)
plt.legend(loc = 5, fontsize=15)
plt.grid(which='major', linestyle=':', color='gray', linewidth= 1)


# Врезка для x -> -inf
xn = -1 * np.arange(1e3, 1e4, 2e-1)
yn = [(xn**bi + ai**bi) / xn**bi for ai, bi in zip(a, b)]

plt.axes([.23, .4, .47, .4])
plt.plot(xn, yn[0], 'r')
plt.plot(xn, yn[1], 'g')
plt.plot(xn, yn[2], 'b')
plt.grid(True)

plt.savefig('Task-1_part-3.svg')
plt.show()
```

Сделайте выводы о поведении графиков при различных значениях $\alpha$ и $\beta$.

Изучите возможности пересечения и то к чему стремятся кривые.

Сделайте выводы о возрастании/убывании.

При увеличении обоих параметров $\alpha$ и $\beta$ функция $f(x; \alpha, \beta) = \frac{{x^\beta}+{\alpha^\beta}}{x^\beta}$ быстрее стремится к своему пределу (очевидно, $\lim_{x \rightarrow \pm\infty}\frac{{x^\beta}+{\alpha^\beta}}{x^\beta}=\lim_{x \rightarrow  \pm\infty}\frac{{x^\beta}}{x^\beta}+\lim_{x \rightarrow  \pm\infty}\frac{{\alpha^\beta}}{x^\beta}=1+0=1:(\forall\alpha,\beta\geq1)$ ), при этом $\beta$ сильнее влияет на быстроту сходимости, чем $\alpha$. При $x>0$ функции вида $f(x; \alpha, \beta=2)$ имеют пересечения с функциям вида $f(x; \alpha, \beta=1)$, при $x<0$ пересечений нет.

## Задача №2

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0$

- $\alpha = 0, \beta = 0$

- $\alpha = 0, \beta = 1$

- $\alpha = 0, \beta = 2$

- $\alpha = 1, \beta = -1$

- $\alpha = 2, \beta = -1$

Сделайте выводы о поведении графиков

```python
import  matplotlib.pyplot  as  plt
import numpy as np

xp = np.arange(0.02, 4, 0.02)
xn = -1 * xp
a = 1, 0, 0, 0, 1, 2
b = 0, 0, 1, 2, -1, -1

yp = [(xp**bi + ai**bi) / xp**bi for ai, bi in zip(a, b)]
yn = [(xn**bi + ai**bi) / xn**bi for ai, bi in zip(a, b)]

plt.figure(figsize=(13,8), dpi=120)
plt.plot(xp, yp[0], 'r', label='$f(x;1,0) = 2$')
plt.plot(xn, yn[0], 'r')
plt.plot(xp, yp[1], 'k', linestyle='--', label='$f(x;0,0) = 2$')
plt.plot(xn, yn[1], 'k', linestyle='--')
plt.plot(xp, yp[2], 'b', label='$f(x;0,1) = \\frac{x}{x}$')
plt.plot(xn, yn[2], 'b')
plt.plot(xp, yp[3], 'y', linestyle='--', label='$f(x;0,2) = \\frac{x^2}{x^2}$')
plt.plot(xn, yn[3], 'y', linestyle='--')
plt.plot(xp, yp[4], 'm', label='$f(x;1,-1) = \\frac{x}{x}+x$')
plt.plot(xn, yn[4], 'm')
plt.plot(xp, yp[5], 'g', label='$f(x;2,-1) = \\frac{x}{x}+\\frac{x}{2}$')
plt.plot(xn, yn[5], 'g')
plt.title('Task 2', fontsize=18)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f(x;\\alpha, \\beta)$', fontsize=15)
plt.legend(loc = 5, fontsize=15)
plt.rc('grid', linestyle=':', color='gray', linewidth= 1)
plt.grid(True)
plt.savefig('Task-2.svg')
plt.show()
```

1. При $\beta=0$ функция вырождается в константу раную 2: $f(x; \alpha, \beta=0)=2$. 
2. При $\alpha=0$ функция вырождается в константу раную 1 (с выколотой точкой $f(x=0)$): $f(x\neq0; \alpha=0, \beta)=2$.
3. При $\beta=-1$ функция вырождается в прямую проходящую через выколотую точку $f(x\neq0; \alpha, \beta=-1)$ с коэффициентом наклона равным $\alpha^{-1}$.

## Задача №3

Построить в общих осях графики для:

- $\alpha = 1, \beta = 0.5$

- $\alpha = 1, \beta = -0.5$

- $\alpha = 1, \beta = -1.5$

Сделайте выводы о поведении графиков, включая возрастание/убывание и выпуклость/вогнутость

```python
xp = np.arange(0.02, 3, 0.02)
a = 1, 1, 1
b = 0.5, -0.5, -1.5

yp = [(xp**bi + ai**bi) / xp**bi for ai, bi in zip(a, b)]

plt.figure(figsize=(13,8), dpi=120)
plt.plot(xp, yp[0], 'r', label='$f(x;1,0.5) = \\frac{x^{0.5}+1}{x^{0.5}}$')
plt.plot(xp, yp[1], 'g', label='$f(x;1,-0.5) = \\frac{x^{-0.5}+1}{x^{-0.5}}$')
plt.plot(xp, yp[2], 'b', label='$f(x;1,-1.5) = \\frac{x^{-1.5}+1}{x^{-1.5}}$')
plt.title('Task 3', fontsize=18)
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$f(x;\\alpha, \\beta)$', fontsize=15)
plt.legend(loc = 5, fontsize=15)
plt.rc('grid', linestyle=':', color='gray', linewidth= 1)
plt.grid(True)
plt.savefig('Task-3.svg')
plt.show()
```

1. При $\beta>0,\alpha=1$ $\lim_{x \rightarrow \pm\infty}\frac{{x^\beta}+{\alpha^\beta}}{x^\beta}=1$, график функции вогнутый.
2.1. При $-1<\beta<0,\alpha=1$ $\lim_{x \rightarrow \pm\infty}\frac{{x^\beta}+{\alpha^\beta}}{x^\beta}=\infty$, график функции выпуклый.
2.2. При $\beta<-1,\alpha=1$ $\lim_{x \rightarrow \pm\infty}\frac{{x^\beta}+{\alpha^\beta}}{x^\beta}=\infty$, график функции вогнутый.
