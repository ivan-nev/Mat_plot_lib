Построить график по заданным точкам

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, c):
    c = list(c)
    if len(c) == 1:
        return c[0]
    return c.pop() * x ** len(c) + f(x, c)


def f_str(c):
    c = list(c)
    if len(c) == 1:
        return str(round(c[0], 3))
    return f'{str(round(c.pop(), 3))}\\cdot x^{str(len(c))} + ' + f_str(c)

# значения для задачи
# mat = np.array(((0,1), (1,1), (2,25), (-1,1), (-2,-23), (0.5,0.90625)), dtype=float)
#Произв пример
mat = np.array([[-3.0, -2.0], [-2.0, 10.0], [-1.0, -15.0], [1.0, -0.5], [4.0, 5.0], [5, -6], [6.0, -7.0], [7, 0], [10, 0],[12,0]])
M = len(mat)

#Ввод своих значений
# M = int(input("Кол-во точек: "))
# mat = np.array([input(f'Координаты точки №{i + 1} в строчку через пробел (x y): ').split() for i in range(M)],dtype=float)
```

```python
def extremum(func, x: np.ndarray, koef, axes, color='ro'):
    derivative = np.gradient(func(x, koef), x)
    extrema_indices = np.where(np.diff(np.sign(derivative)))
    extrema_x = x[extrema_indices]
    extrema_y = func(x[extrema_indices], koef)
    axes.plot(extrema_x, extrema_y, color, label='Экстремумы')
    axes.tick_params(axis='x', rotation=70)
    axes.set_xticks(np.around(np.concatenate([np.linspace(x[0], x[-1], 2), extrema_x, mat[:,0]]), 2))
    axes.set_yticks(np.around(np.concatenate([extrema_y, mat[:,1]]), 2))
```

```python
k = mat[:, 0]
v1 = mat[:, -1]
m1 = np.array(tuple(map(lambda x: [x ** i for i in range(M)], k)))


if np.linalg.det(m1):
    koef = np.linalg.solve(m1, v1)

x = np.linspace(min(k), max(k), 500)
# x = np.linspace(-100,100, 500)
fig, ax = plt.subplots(1, figsize=(15, 10))
ax.plot(x, f(x, koef), label='F(x)')
ax.plot(mat[:, 0], mat[:, 1], 'ro', label='Заданные точки')
ax.set_title(f'$F(x) = {f_str(koef)}$')
ax.grid(which='major', linestyle=':', color='#aaa')
extremum(f, x, koef, ax, color='bo')
ax.legend()
# ax.set_yscale('symlog', base=20)
# ax.set_xlim(-0.6,0.6)
# ax.set_ylim(0,1.4)
plt.show()
```
